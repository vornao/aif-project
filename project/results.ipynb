```python
import pandas as pd 
import numpy as np

import csv
import json
import seaborn as sns
import matplotlib.pyplot as plt

# Example JSON data (replace this with your actual JSON data)
with open('/Users/vornao/Developer/aif-project/project/winners_16_irene.json') as json_file:
    json_data = json.load(json_file)

csv_file = "output.csv"

# Open the CSV file in write mode
with open(csv_file, 'w', newline='') as csvfile:
    # Create a CSV writer object
    csv_writer = csv.writer(csvfile)

    # Write the header (assuming keys of the first JSON object as header)
    header = json_data[0].keys()
    csv_writer.writerow(header)

    # Write each JSON object as a row in the CSV file
    for item in json_data:
        csv_writer.writerow(item.values())
```

```python
df = pd.read_csv('output.csv')
df
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(15, 5))
sns.histplot(data=df['best_fitness'], ax=axs[1], kde=True)
sns.histplot(data=df['generation'], ax=axs[0], kde=True)
# set title: distribution of best fitness and generation, 16 individuals per generation, map: irene.des
fig.suptitle('Distribution of best fitness and generation, 16 individuals per generation, map: irene.des')
plt.show()
```
